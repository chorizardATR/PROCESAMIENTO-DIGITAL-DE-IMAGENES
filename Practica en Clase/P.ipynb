```python
import numpy as np
import matplotlib.pyplot as plt

size = 400
y, x = np.ogrid[-size/2:size/2, -size/2:size/2]

radius = np.sqrt(x**2 + y**2) 

ancho_anillo = 40
diana = (radius // ancho_anillo) % 2

# Graficar
plt.figure(figsize=(5, 5))
plt.imshow(diana, cmap='gray')
plt.axis('off')
plt.title('Diana (Tiro al blanco)')
plt.show()
```
