```python
import cv2
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [20, 8]

img_gray = cv2.imread('colores.jpg', cv2.IMREAD_GRAYSCALE)


plt.subplot(2, 5, 1)
plt.title('Falso Color: viridis')
plt.imshow(img_gray, cmap='viridis')
plt.axis('off')

plt.subplot(2, 5, 2)
plt.title('Falso Color: plasma')
plt.imshow(img_gray, cmap='plasma')
plt.axis('off')

plt.subplot(2, 5, 3)
plt.title('Falso Color: inferno')
plt.imshow(img_gray, cmap='inferno')
plt.axis('off')

plt.subplot(2, 5, 4)
plt.title('Falso Color: magma')
plt.imshow(img_gray, cmap='magma')
plt.axis('off')

plt.subplot(2, 5, 5)
plt.title('Falso Color: cividis')
plt.imshow(img_gray, cmap='cividis')
plt.axis('off')





plt.subplot(2, 5, 6)
plt.title('Falso Color: jet')
plt.imshow(img_gray, cmap='jet')
plt.axis('off')

plt.subplot(2, 5, 7)
plt.title('Falso Color: hot')
plt.imshow(img_gray, cmap='hot')
plt.axis('off')

plt.subplot(2, 5, 8)
plt.title('Falso Color: ocean')
plt.imshow(img_gray, cmap='ocean')
plt.axis('off')

plt.subplot(2, 5, 9)
plt.title('Falso Color: cool')
plt.imshow(img_gray, cmap='cool')
plt.axis('off')

plt.subplot(2, 5, 10)
plt.title('Falso Color: hsv')
plt.imshow(img_gray, cmap='hsv')
plt.axis('off')


plt.tight_layout() 
plt.show()
```
