```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def solarizar(imagen, umbral):
    # Aplicamos solarización
    img_sol = np.where(imagen < umbral, imagen, 255 - imagen)
    return img_sol

imagen = cv2.imread('colores_nuevo.jpg')
img_gray = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

umbral = 127
img_fin = solarizar(img_gray, umbral)

plt.figure(figsize=(12,6))

# Imagen original
plt.subplot(1,2,1)
plt.title("Imagen Original")
plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Imagen solarizada
plt.subplot(1,2,2)
plt.title("Imagen Solarizada")
plt.imshow(img_fin, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def posterizar (imagen, niveles):
    factor_escala= 256 // niveles
    img_pos=np.round(imagen / factor_escala) * factor_escala
    return img_pos.astype(np.uint8)
def solarizar(imagen, umbral):
    # Aplicamos solarización
    img_sol = np.where(imagen < umbral, imagen, 255 - imagen)
    return img_sol
imagen = cv2.imread('colores_nuevo.jpg')
umbral = 127
img_sol = solarizar(imagen, umbral)

nvl = 1.5
img_pos = posterizar(imagen, nvl)

plt.figure(figsize=(18,6))

plt.subplot(1,3,1)
plt.title('Original')
plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))

plt.subplot(1,3,2)
plt.title('Solarizada')
plt.imshow(cv2.cvtColor(img_sol, cv2.COLOR_BGR2RGB))

plt.subplot(1,3,3)
plt.title('Posterizada')
plt.imshow(cv2.cvtColor(img_pos, cv2.COLOR_BGR2RGB))

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def falso_color (img_gray):
    escal_img=cv2.normalize(img_gray, None, 0, 255, cv2.NORM_MINMAX)
    falso_img=cv2.applyColorMap(escal_img, cv2.COLORMAP_JET)
    return falso_img

imagen=cv2.imread('colores_nuevo.jpg')
img_gray=cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
img_fin=falso_color(img_gray)
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)  
plt.imshow(cv2.cvtColor(img_fin, cv2.COLOR_BGR2RGB))
plt.subplot(1,2,2)
plt.imshow(img_gray, cmap='gray')
plt.show()
```
